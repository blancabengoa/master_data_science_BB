## 1. Importamos librerías necesarias

```python
import pandas as pd
import numpy as np
import datetime
```

## 2. Lectura y testeo de datos

```python
df=pd.read_csv('../data/EMBALSES.csv',sep=';')
df.head(12)
```

```python
df.dtypes
```

Los meses no me los reconoce con el formato %b al estar en español, así que los convierto a código numérico a través de un diccionario

```python
calendar=list(['ene','feb','mar','abr','may','jun','jul','ago','sept','oct','nov','dic'])
dict={}
for m in range(12):
    dict[calendar[m]]=m+1
    
aux=df['MES'].str.split('-') 
fecha=[]
for i in aux:
    mes=dict[i[0]]
    ano=i[1]
    fecha.append(str(mes)+'-'+str(ano))
    #fecha.append(pd.to_datetime(datetime.datetime.strptime(str(mes)+'-'+i[1],'%m-%y')))
    
df['mes_ano']=pd.Series(fecha)

df.head()
```

Los datos extraídos son mensuales, todos mis datos son horarios. De momento convierto todos los datos a diarios y, cuando unifique todos en un único archivo, los pasaré a horarios.

Para pasarlos a horarios simplemente pondré el mismo valor todos los días del mes ya que no tendré el dato del estado del embalse para el día siguiente con mayor precisión en la vida real.

```python
embalses=pd.DataFrame(pd.date_range(start='2015-10-01',end='2018-09-30'),columns=['datetime'])
embalses.head()
```

```python
#me creo una columna aux mes_ano igual que la anterior para unir después ambos dataframes por esta columna
embalses['mes_ano']=[str(x.month) + '-' + str(x.year)[-2:] for x in embalses['datetime']] 
embalses.head()
```

```python
embalses_dia=embalses.merge(df,on='mes_ano')
embalses_dia.head(32)
```
